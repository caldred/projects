```python
import glob
import numpy as np
import csv
import matplotlib.pyplot as plt
import pandas as pd
import joblib
```

```python
directory = 'C:\\Users\\Cal\\Documents\\PioneerDecklists'
cards = {}
card_first = {}
n_cards = 0
n_decks = 0
for filename in glob.iglob(directory+'\\*\\*.txt'):
    deck = np.loadtxt(filename, usecols=(0,), delimiter = '\n', dtype="str")
    n_decks += 1
    print(filename)
    for card in deck:
        name = card.split(sep = " ", maxsplit = 1)[1]
        if name not in cards:
            print(name, n_cards)
            cards[name] = n_cards
            card_first[name] = n_decks
            n_cards += 1
```

```python
for key in card_first:
    card_first[key] /= n_decks
    card_first[key] = 2/(2-card_first[key])
```

```python
banned_cards = [
    "Bloodstained Mire",
    "Felidar Guardian",
    "Field of the Dead",
    "Flooded Strand",
    "Inverter of Truth",
    "Kethis, the Hidden Hand",
    "Leyline of Abundance",
    "Nexus of Fate",
    "Oko, Thief of Crowns",
    "Once Upon a Time",
    "Polluted Delta",
    "Smuggler's Copter",
    "Underworld Breach",
    "Veil of Summer",
    "Walking Ballista",
    "Windswept Heath",
    "Wooded Foothills",
]
```

```python
banned_indices = []
for card in banned_cards:
    if card in cards:
        print(card, cards.get(card))
        banned_indices.append(cards[card])
```

```python
decks = {}
banned_decks= {}
for filename in glob.iglob(directory+'\\*\\*.txt'):
    deck = np.loadtxt(filename, usecols=(0,), delimiter = '\n', dtype="str")
    deck_num = int(filename[-11:-4])
    print(deck_num)
    decks[deck_num] = np.zeros(n_cards)
    for card in deck:
        n = card.split(sep = " ", maxsplit = 1)[0]
        name = card.split(sep = " ", maxsplit = 1)[1]
        decks[deck_num][cards[name]] = n
        if cards[name] in banned_indices:
            banned_decks[deck_num] = True
```

```python
tot = np.zeros(n_cards)
for key in decks:
    tot += decks[key]
```

```python
rev_cards = {}
for key in cards:
    rev_cards[cards[key]] = key
```

```python
card_tot = {}
for key in rev_cards:
    card_tot[rev_cards[key]] = tot[key]
```

```python
common_cards = {}
n100 = 0
for key in card_tot:
    if card_tot[key] > 99 and key not in banned_cards:
        n100+=1
        common_cards[key] = n100
print(n100)
```

```python
for key in common_cards:
    print(key)
```

```python
decks = {}
banned_decks= {}
for filename in glob.iglob(directory+'\\*\\*.txt'):
    deck = np.loadtxt(filename, usecols=(0,), delimiter = '\n', dtype="str")
    deck_num = int(filename[-11:-4])
    print(deck_num)
    decks[deck_num] = np.zeros(n100+1)
    for card in deck:
        n = card.split(sep = " ", maxsplit = 1)[0]
        name = card.split(sep = " ", maxsplit = 1)[1]
        decks[deck_num][common_cards.get(name, 0)] += int(n)
        if name in banned_cards:
            banned_decks[deck_num] = True
```

```python
valid_decks = {}
for deck in decks:
    if deck in banned_decks:
        continue
    else:
        valid_decks[deck] = decks[deck]
```

```python
df = pd.DataFrame.from_dict(valid_decks, orient='index')
```

```python
msk = np.random.rand(len(df)) < 0.9
joblib.dump(df[msk], 'data/common_train.dat')
joblib.dump(df[~msk], 'data/common_test.dat')
```
