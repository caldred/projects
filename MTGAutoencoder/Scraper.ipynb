```python
import requests
from bs4 import BeautifulSoup
import time
import os
import numpy as np

deck_dir = 'C:/Users/Cal/Documents/PioneerDecklists/'

def scrape(url):
    resp = requests.get(url)
    
    if resp.status_code == 200:
        print("Successfully opened the web page")
        
        soup = BeautifulSoup(resp.text,'html.parser')
        return soup
    else:
        print('Error')
        return -1
```

```python
tourneys = []
page = 1
while True:
    print(page)
    search_url = f'https://www.mtggoldfish.com/tournament_searches/create?commit=Search&page={page}&tournament_search%5Bdate_range%5D=10%2F04%2F2009+-+10%2F04%2F2029&tournament_search%5Bformat%5D=pioneer&tournament_search%5Bname%5D=&utf8=%E2%9C%93'
    page += 1
    pioneer = scrape(search_url)
    tr = pioneer.find_all('tr')
    if len(tr) == 0:
        break
    for r in tr:
        td_list = r.find_all('td')
        try:
            suffix = td_list[1].find('a').get('href')
            suffix = suffix[-5:]
            print(int(suffix))
        except:
            print('No href')
        else:
            print(suffix)
            url = 'https://www.mtggoldfish.com/tournament/'+suffix
            tourneys.append(suffix)
```

```python
t = [x[1] for x in os.walk(deck_dir)][0]
t = list(map(int, t))
max_tourney = max(t)

for tourney in tourneys:
    if int(tourney) <= max_tourney:
        continue
    soup = scrape('https://www.mtggoldfish.com/tournament/'+tourney)
    if str(soup) == 'Throttled':
        break
    os.mkdir(deck_dir + tourney)
    tr = soup.find_all('tr')
    for r in tr:
        td_list = r.find_all('td')
        try:
            suffix = td_list[1].find('a').get('href')
            if len(suffix) > 25:
                raise ValueError('suffix too long')
            suffix = suffix[-7:]
            suffix = str(int(suffix))
        except:
            print('No href')
        else:
            time.sleep(np.random.random()*2)
            print(suffix)
            url = 'https://www.mtggoldfish.com/deck/download/'+suffix
            resp = requests.get(url)
            deck = BeautifulSoup(resp.text,'html.parser')
            deck = str(deck)
            deck = deck.replace('\r', '')
            f = open(deck_dir + tourney + '/' + suffix +'.txt', "w+")
            f.write(str(deck))
            f.close()
```
