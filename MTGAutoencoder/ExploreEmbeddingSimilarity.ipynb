```python
import glob
import numpy as np
import csv
import matplotlib.pyplot as plt
import pandas as pd
import joblib
```

```python
a = np.loadtxt('data/encoder_embedding.csv', delimiter=',')
b = np.loadtxt('data/decoder_embedding.csv', delimiter=',')
c = joblib.load('data/common_inv.dat')
```

```python
dot = {}
for i in range(len(a)):
    print(c.get(i, 'Other'))
    dot[i] = np.dot(a, a[i])
    ind = np.argpartition(dot[i], -15)[-15:]
    for j in ind:
        print(j, c.get(j, 'Other'), dot[i][j])
    print()
```
