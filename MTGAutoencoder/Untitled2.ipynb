```python
import requests
from bs4 import BeautifulSoup
import time
import os
import numpy as np

def scrape(url):
    resp = requests.get(url)
    
    if resp.status_code == 200:
        print("Successfully opened the web page")
        
        soup = BeautifulSoup(resp.text,'html.parser')
        return soup
    else:
        print('Error')
        return -1
```

```python
soup = scrape('https://www.mtggoldfish.com/tournament/pioneer-league-2020-06-29')
```

```python
decks = {}
tr = soup.find_all('tr')
for r in tr:
    td_list = r.find_all('td')
    try:
        suffix = td_list[1].find('a').get('href')
        suffix = suffix[:5]+'/download'+suffix[5:]
        if len(suffix) > 25:
            raise ValueError('suffix too long')
    except:
        print('No href')
    else:
        time.sleep(.200)
        print(suffix)
        url = 'https://www.mtggoldfish.com'+suffix
        resp = requests.get(url)
        deck = BeautifulSoup(resp.text,'html.parser')
        decks.append(deck)
```

```python
tourneys = []
page = 1
while True:
    print(page)
    search_url = f'https://www.mtggoldfish.com/tournament_searches/create?commit=Search&page={page}&tournament_search%5Bdate_range%5D=10%2F04%2F2009+-+10%2F04%2F2029&tournament_search%5Bformat%5D=pioneer&tournament_search%5Bname%5D=&utf8=%E2%9C%93'
    page += 1
    pioneer = scrape(search_url)
    tr = pioneer.find_all('tr')
    if len(tr) == 0:
        break
    for r in tr:
        td_list = r.find_all('td')
        try:
            suffix = td_list[1].find('a').get('href')
            suffix = suffix[-5:]
            print(int(suffix))
        except:
            print('No href')
        else:
            print(suffix)
            url = 'https://www.mtggoldfish.com/tournament/'+suffix
            tourneys.append(suffix)
```
